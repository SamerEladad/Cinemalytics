```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as mp
import seaborn as sns
```

```python
# Clean combined movies dataset
movies_clean_path ='updated_clean_combined_movies.csv'
movies_clean_df = pd.read_csv(movies_clean_path)

# Display Clean Movies dataset column names
print("\nClean Combined Movies dataset column names:")
print(movies_clean_df.columns.tolist())
```

```python
print(movies_clean_df.head())
```

# **Correlation Between Movie Count and Age Rating**

```python
# Count the number of movies for each age rating
rating_counts = movies_clean_df['age_rating'].value_counts()

# Plotting with salmon color
plt.figure(figsize=(10, 6))
rating_counts.plot(kind='bar', color='salmon')
plt.title('Number of Movies by Age Rating')
plt.xlabel('Age Rating')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# **Correlation Between Movie Count and Genre**

```python
# Count the number of movies for each genre
genre_counts = movies_clean_df['genre'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
genre_counts.plot(kind='pie', explode=[0.1] * len(genre_counts), autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Movies by Genre')
plt.ylabel('')  # Remove y-label
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show()
```

# **Correlation Between Movie Count and Duration**

```python
# Count the number of movies for each duration
duration_counts = movies_clean_df['duration_min'].value_counts()

# Plotting as a dot plot with salmon color
plt.figure(figsize=(10, 6))
plt.scatter(duration_counts.index, duration_counts.values, color='salmon', s=50, alpha=0.7)

# Adding title and labels
plt.title('Distribution of Movies by Duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

# Adding gridlines for the y-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjusting layout
plt.tight_layout()

# Display the plot
plt.show()
```

# **Correlation Between Movie Count and Platform**

```python
# Count the number of movies for each age rating
platform_counts = movies_clean_df['platform'].value_counts()

# Plotting with different colors for each platform
plt.figure(figsize=(10, 6))
platform_counts.plot(kind='bar', color=plt.cm.tab10.colors)
plt.title('Number of Movies by Platform')
plt.xlabel('Platform')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# **Correlation Between the Age Rating and Revenue**

```python
# Function to create a scatter plot showing the relationship between movie duration and revenue for a specific platform
def plot_duration_vs_revenue_for_platform(df, platform_name):
    plt.figure(figsize=(14, 8))
    sns.scatterplot(data=df, x='duration_min', y='revenue', color='salmon', alpha=0.5)
    plt.title(f'Relationship Between Movie Duration and revenue on {platform_name}', fontsize=16)
    plt.xlabel('Duration (minutes)', fontsize=14)
    plt.ylabel('revenue', fontsize=14)
    plt.grid(True)
    plt.tight_layout()  # Adjust layout to make sure everything fits without overlapping
    plt.show()

# Assuming 'movies_clean_df' is your DataFrame containing movie data
plot_duration_vs_revenue_for_platform(movies_clean_df, 'all platforms')
```

# **Correlation Between the Age Rating and Budget**

```python
# Define a palette of colors for each age rating category
palette = {'G': 'skyblue', 'PG': 'salmon', 'PG-13': 'gold', 'R': 'lightgreen', 'NC-17': 'orchid'}

# Get unique age rating categories from the DataFrame
unique_age_ratings = movies_clean_df['age_rating'].unique()

# Create a palette with default color for missing keys
default_color = 'gray'
custom_palette = {rating: palette.get(rating, default_color) for rating in unique_age_ratings}

# Plotting as a violin plot with hue assigned to 'age_rating'
plt.figure(figsize=(10, 6))
sns.violinplot(data=movies_clean_df, x='age_rating', y='budget', hue='age_rating', palette=custom_palette, alpha=0.5, legend=False)
plt.title('Relation Between Age Rating and Budget')
plt.xlabel('Age Rating')
plt.ylabel('Budget')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
